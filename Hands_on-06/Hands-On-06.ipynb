```python
import time
import random
import matplotlib.pyplot as plt
def iterative_quicksort(arr):
    stack = [(0, len(arr) - 1)]

    while stack:
        smaller_than_pivot, greater_than_pivot = stack.pop()
        if smaller_than_pivot < greater_than_pivot:
            pivot = partition(arr, smaller_than_pivot, greater_than_pivot)
            stack.append((smaller_than_pivot, pivot - 1))
            stack.append((pivot + 1, greater_than_pivot))

def partition(arr, smaller_than_pivot, greater_than_pivot):
    pivot = arr[smaller_than_pivot]
    i = smaller_than_pivot + 1
    j = greater_than_pivot

    while True:
        while i <= j and arr[i] <= pivot:
            i += 1
        while i <= j and arr[j] >= pivot:
            j -= 1
        if i <= j:
            arr[i], arr[j] = arr[j], arr[i]
        else:
            break

    arr[smaller_than_pivot], arr[j] = arr[j], arr[smaller_than_pivot]
    return j

def benchmark_for_quicksort():
    array_sizes = [100, 500, 1000, 5000, 10000]
    best_case_times = []
    worst_case_times = []
    average_case_times = []

    for size in sizes:
        best_case = list(range(array_sizes))
        start_time = time.time()
        iterative_quicksort(best_case.copy())
        best_case_times.append(time.time() - start_time)

        worst_case = list(range(array_sizes, 0, -1))
        start_time = time.time()
        iterative_quicksort(worst_case.copy())
        worst_case_times.append(time.time() - start_time)

        average_case = [random.randint(0, array_sizes) for _ in range(array_sizes)]
        start_time = time.time()
        iterative_quicksort(average_case.copy())
        average_case_times.append(time.time() - start_time)

    plt.plot(array_sizes, best_case_times, label="Best Case")
    plt.plot(array_sizes, worst_case_times, label="Worst Case")
    plt.plot(array_sizes, average_case_times, label="Average Case")
    plt.xlabel("Input Size in terms of N")
    plt.ylabel("Time in seconds")
    plt.legend()
    plt.title("Non-Random Pivot Quicksort Performance")
    plt.show()

# Run benchmark
benchmark_iterative_quicksort()
```
