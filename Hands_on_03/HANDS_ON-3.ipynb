```python
import time
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

# Original function given
def original_function(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

#for various n values
ns = np.array(range(1, 1001, 50))  # Test from n = 1 to 1000
original_time = []

for n in ns:
    start = time.time()
    original_function(n)
    end = time.time()
    original_time.append(end - start)

# Fit a polynomial to the timings (quadratic curve)
coeffs = np.polyfit(ns, original_time, 2)
p = Polynomial(coeffs[::-1])  # Polynomial object

# Plot time vs n
plt.plot(ns, original_time, 'bo-', label='Original Function')
plt.plot(ns, p(ns), 'r-', label='Fitted Polynomial Curve')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Time vs n for Original Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Defining upper and lower bounds
upper_bound = lambda n: 1.2 * p(n)
lower_bound = lambda n: 0.8 * p(n)

# Plot with upper and lower bounds
plt.plot(ns, original_time, 'bo-', label='Original Function')
plt.plot(ns, p(ns), 'r-', label='Fitted Polynomial Curve')
plt.plot(ns, upper_bound(ns), 'p--', label='Upper Bound')
plt.plot(ns, lower_bound(ns), 'v--', label='Lower Bound')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Time vs n with Upper and Lower Bounds')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# Adding a vertical line at n_0
n_0 = 200  # Eyeballed based on the plot
plt.plot(ns, original_time, 'bo-', label='Original Function')
plt.plot(ns, p(ns), 'r-', label='Fitted Polynomial Curve')
plt.axvline(x=n_0, color='yellow', linestyle='--', label='n_0 = {}'.format(n_0))
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Time vs n with n_0 Indicated')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#code for merge sort
def merge_sort(arr):
    if len(arr) > 1:
        #  middle point and split the array
        middle = len(arr) // 2
        l_half = arr[:middle]
        r_half = arr[middle:]

        #  sort the left and right half recursively
        merge_sort(l_half)
        merge_sort(r_half)

        # Merging array
        a = b = c = 0
        print(f"Left: {l_half}, Right: {r_half}")  
        while a < len(l_half) and b < len(r_half):
            if l_half[a] < r_half[b]:
                arr[c] = l_half[a]
                a += 1
            else:
                arr[c] = r_half[b]
                b += 1
            c += 1

        # Check if any element was left
        while a < len(l_half):
            arr[c] = l_half[a]
            a += 1
            c += 1

        while b < len(r_half):
            arr[c] = r_half[b]
            b += 1
            c += 1

        print(f"Merged: {arr}")  # Show merging of arrays

# Implement merge sort on the given array
test_array = [5, 2, 4, 7, 1, 3, 2, 6]
print("Original array:", test_array)
merge_sort(test_array)
print("Sorted array:", test_array)
```
