```python
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bblSort(array):
    no_of_elements = len(array)
    for i in range(no_of_elements):
        swap = False
        for j in range(0, no_of_elements-i-1):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
                swap = True
        if not swap:
            break

def insertionsort(array): 
    for i in range(1, len(array)): 
        element = array[i] 
        j = i-1 
        while j >= 0 and element < array[j]: 
            array[j + 1] = array[j] 
            j -= 1 
        array[j + 1] = element 

def selectionsort(array):
    for i in range(len(array)):
        min_index = i
        for j in range(i+1, len(array)):
            if array[j] < array[min_index]:
                min_index = j
        array[i], array[min_index] = array[min_index], array[i]

sorts = [ 
    { 
        "name": "Bubble Sort", 
        "sort": lambda array: bblSort(array) 
    },
    { 
        "name": "Insertion Sort", 
        "sort": lambda array: insertionsort(array) 
    }, 
    { 
        "name": "Selection Sort", 
        "sort": lambda array: selectionsort(array) 
    }, 
]   

plt.xlabel('Length of Array Elements') 
plt.ylabel('Time Complexity') 

#For 1-100 values
list_size = np.array([i for i in range(1, 100)]) 

for sorting in sorts: 
    timelist = list() 
    start_all = time.perf_counter() 
    for i in range(1, 100): 
        start = time.perf_counter() 
        a = np.random.randint(1000, size=i)          
        sorting["sort"](a) 
        end = time.perf_counter() 
        timelist.append(end - start) 
    end_all = time.perf_counter() 
    plt.plot(list_size, timelist, label=sorting["name"]) 
plt.grid() 
plt.legend() 
plt.show()
```
